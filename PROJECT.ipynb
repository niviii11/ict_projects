1.Text-Based Decision Game

Create a text-based decision-making game. The game should present players with scenarios where choices affect the storyline and score. Multiple story paths and endings must exist. The system should track decisions, compute scores, and allow replayability. String processing and conditional branching are core requirements.

```python
def play_game():
    score = 0
    decisions = []

    print("\nTHE LOST EXPEDITION ")
    name = input("Enter your name: ").strip().title()
    print(f"\nWelcome, {name}! Your journey begins...\n")
    print("You are lost in a dense forest.")
    print("1. Follow a narrow path")
    print("2. Walk towards the river")

    choice = input("Enter choice (1/2): ")
    if choice == "1":
        score += 10
        decisions.append("Path")
        print("You safely reach a small village.")
    else:
        score -= 5
        decisions.append("River")
        print("You slip and lose some supplies.")

    print("\nNight falls. You need shelter.")
    print("1. Build a shelter")
    print("2. Keep walking in the dark")

    choice = input("Enter choice (1/2): ")
    if choice == "1":
        score += 15
        decisions.append("Shelter")
        print("You rest well and recover strength.")
    else:
        score -= 10
        decisions.append("Walk")
        print("You get injured in the dark.")

    print("\nYou encounter a wild animal.")
    print("1. Fight")
    print("2. Run")

    choice = input("Enter choice (1/2): ")
    if choice == "1":
        score -= 15
        decisions.append("Fight")
        print("You survive but are badly wounded.")
    else:
        score += 10
        decisions.append("Run")
        print("You escape safely.")

    print("\n--- GAME OVER ---")
    print("Decisions Taken:", " -> ".join(decisions))
    print("Final Score:", score)

    if score >= 25:
        print("ENDING: HEROIC SURVIVOR")
    elif score >= 10:
        print("ENDING: LUCKY ESCAPE")
    else:
        print("ENDING: LOST FOREVER")

def main_menu():
    while True:
        print("\n--- TEXT-BASED DECISION GAME ---")
        print("1. Start Game")
        print("2. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            play_game()
        elif choice == "2":
            print("Thank you for playing!")
            break
        else:
            print("Invalid choice. Try again.")

main_menu()
```

2. Quiz Engine with Dynamic Scoring

Create a Quiz Engine that dynamically evaluates participants. The system should manage a question bank categorized by difficulty levels. During quiz attempts, questions should appear in a pseudo-random order without using random libraries. Scoring must vary by difficulty, include negative marking, and generate detailed score reports. String handling and nested data structures must be used extensively.

```python
questions = {
    "Easy": [
        ("What is 2 + 2?", "4"),
        ("Capital of India?", "Delhi")
    ],
    "Medium": [
        ("What is 12 * 8?", "96"),
        ("Who developed Python?", "Guido")
    ],
    "Hard": [
        ("Binary of 10?", "1010"),
        ("Year Python was released?", "1991")
    ]
}

scores = {
    "Easy": (1, 0.5),
    "Medium": (2, 1),
    "Hard": (3, 1.5)
}


def pseudo_shuffle(lst):
    new = []
    i = len(lst) - 1
    while i >= 0:
        new.append(lst[i])
        i -= 1
    return new

def start_quiz():
    total_score = 0
    report = []

    for level in questions:
        q_list = pseudo_shuffle(questions[level])

        for q, ans in q_list:
            print("\n(", level, ") ", q)
            user_ans = input("Answer: ").strip()

            if user_ans.lower() == ans.lower():
                total_score += scores[level][0]
                report.append((q, "Correct", scores[level][0]))
            else:
                total_score -= scores[level][1]
                report.append((q, "Wrong", -scores[level][1]))

    print("\n--- QUIZ REPORT ---")
    for r in report:
        print("Q:", r[0], "|", r[1], "| Score:", r[2])

    print("Final Score:", total_score)

while True:
    print("\nQUIZ ENGINE")
    print("1. Start Quiz")
    print("2. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        start_quiz()
    elif choice == "2":
        print("Quiz Ended.")
        break
    else:
        print("Invalid choice.")
```
